```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Set up the visual theme
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
#Load the dataset
file_path = "ipl.matches.csv"
ipl_df = pd.read_csv(file_path)
```

```python
#Convert 'date' to datetime
ipl_df['date'] = pd.to_datetime(ipl_df['date'])
```

```python
#Fill missing values
ipl_df['city'].fillna('Unknown', inplace=True)
ipl_df['player_of_match'].fillna('Unknown', inplace=True)
ipl_df['winner'].fillna('No Result', inplace=True)
ipl_df['result_margin'].fillna(0, inplace=True)
ipl_df['target_runs'].fillna(0, inplace=True)
ipl_df['target_overs'].fillna(0, inplace=True)
ipl_df['method'].fillna('Normal', inplace=True)
```

```python
#Drop duplicates
ipl_df.drop_duplicates(inplace=True)
```

```python
#Feature Engineering: Did toss winner also win match?
ipl_df['toss_match_win'] = ipl_df['toss_winner'] == ipl_df['winner']
```

```python
#1.Matches per season
plt.figure()
sns.countplot(data=ipl_df, x='season', order=ipl_df['season'].value_counts().index)
plt.title("Number of Matches per Season")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("1_matches_per_season.png")
plt.show()
```

```python
#2.Top 10 Winning Teams
plt.figure()
sns.countplot(data=ipl_df, y='winner', order=ipl_df['winner'].value_counts().head(10).index)
plt.title("Top 10 Winning Teams")
plt.tight_layout()
plt.savefig("2_top_winning_teams.png")
plt.show()
```

```python
#3.Toss Decision Distribution
plt.figure()
sns.countplot(data=ipl_df, x='toss_decision')
plt.title("Toss Decision Distribution")
plt.tight_layout()
plt.savefig("3_toss_decision.png")
plt.show()
```

```python
#4.Toss Winner vs Match Winner
plt.figure()
sns.countplot(x='toss_match_win', data=ipl_df)
plt.title("Did Toss Winner Also Win the Match?")
plt.xticks([0, 1], ['No', 'Yes'])
plt.tight_layout()
plt.savefig("4_toss_match_win.png")
plt.show()
```

```python
#5.Top 10 Cities by Match Count
plt.figure()
sns.countplot(y='city', data=ipl_df, order=ipl_df['city'].value_counts().head(10).index)
plt.title("Top 10 Cities by Number of Matches")
plt.tight_layout()
plt.savefig("5_top_cities.png")
plt.show()
```

```python
#6.Match Result Types
plt.figure()
sns.countplot(x='result', data=ipl_df)
plt.title("Match Result Types")
plt.tight_layout()
plt.savefig("6_result_types.png")
plt.show()
```

```python
#7.Super Over Matches
plt.figure()
sns.countplot(x='super_over', data=ipl_df)
plt.title("Super Over Matches")
plt.tight_layout()
plt.savefig("7_super_over.png")
plt.show()
```

```python
#8.Distribution of Victory Margins (Runs)
plt.figure()
sns.histplot(ipl_df[ipl_df['result'] == 'runs']['result_margin'], bins=30, kde=True)
plt.title("Distribution of Victory Margins (Runs)")
plt.tight_layout()
plt.savefig("8_margin_runs.png")
plt.show()
```

```python
#9.Distribution of Victory Margins (Wickets)
plt.figure()
sns.histplot(ipl_df[ipl_df['result'] == 'wickets']['result_margin'], bins=10, kde=True)
plt.title("Distribution of Victory Margins (Wickets)")
plt.tight_layout()
plt.savefig("9_margin_wickets.png")
plt.show()
```

```python
#10.Top 10 Players of the Match
plt.figure()
sns.countplot(y='player_of_match', data=ipl_df, order=ipl_df['player_of_match'].value_counts().head(10).index)
plt.title("Top 10 Player of the Match Winners")
plt.tight_layout()
plt.savefig("10_player_of_match.png")
plt.show()
```
